```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import arange,sin,cos 
```

```python
def f(alpha, beta, a):
       return ( 2 * alpha * beta * cos(alpha*a) + (pow(beta, 2)- pow(alpha, 2)) * sin(alpha*a))
```

```python
diccio = dict()
```

```python
u = 1
a = 1
stepsize=1e-3
Energy = arange(0.0, u, stepsize)
alpha = np.sqrt(2 * Energy)
beta = np.sqrt(-2*(Energy - u))

for valordea in range(5,11):
   
    resultado = f(alpha,beta,valordea)
    energias=np.where(abs(resultado)<1e-2)[0]
    plt.figure(valordea)
    plt.title('Energía en pozo de potencial', fontsize=20, fontweight='bold')
    fn = plt.scatter(Energy, resultado, color='red')
    plt.scatter(energias*stepsize, resultado[energias], color='blue', marker='s')
    plt.xlabel('Energía', fontsize=15)
    plt.yticks(fontsize=10)
    plt.xticks(fontsize=10)
    plt.ylabel('Función de valores propios', fontsize=15)
    print('Los valores propios de energía son', stepsize*energias[1:])   
    
    diccio[valordea]=stepsize*energias[1:]
```
